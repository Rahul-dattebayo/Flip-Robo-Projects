## Q1. Scrape the details of most viewed videos on YouTube from Wikipedia:

#### Importing required libraries

```python
import pandas as pd
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

#### Opening the desired website

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

#### Initialising the respective cell values to form a table

```python
#1.Rank

rank = []

for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[1]'):
    rank.append(i.text.split('.')[0])
    
rank

#video_name

video_name = []

for j in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]'):
    video_name.append(j.text.split('[')[0])
    
len(video_name)

#uploader

uploader = []

for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]'.format(i)):
    uploader.append(i.text)
    
uploader

#views

views = []

for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]'.format(i)):
    views.append(i.text)
    
views

#date

date = []

for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[5]'.format(i)):
    date.append(i.text)
```

```python
print(len(rank), len(video_name), len(uploader), len(date), len(views))
```

#### DataFraming the data

```python
table = pd.DataFrame({})
table['Rank'] = rank
table['Name'] = video_name
table['Artist'] = uploader
table['Upload date'] = date
table['Views(Billions)'] = views
```

```python
table
```
