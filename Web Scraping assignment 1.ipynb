### Write a python program to display IMDB’s Top rated 100 movies’ data 
(i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
# Creating the lists we want to write into
titles = []
years = []
imdb_ratings = []
```

```python
import numpy as np
count = np.arange(1, 101, 1)
count
```

```python
# Storing each of the urls of 100 movies 

for i in count:
    # Getting the contents from the url
    i = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc' + str(i) + '&ref_=adv_nxt')
    soup2 = BeautifulSoup(i.text, 'html.parser')
    
    # Aiming the part of the html we want to get the information from
    movie_div = soup2.find_all('div', class_='lister-item mode-advanced')
    
    
    for container in movie_div:
        # Scraping the movie's name
        name = container.h3.a.text
        titles.append(name)
        
        # Scraping the movie's year
        year = container.h3.find('span', class_='lister-item-year').text
        years.append(year)
        
        # Scraping the rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
```

```python
import pandas as pd

movies = pd.DataFrame({'movie':titles,
                       'year':years,
                       'imdb_rating':imdb_ratings})

movies.head(100)
```

```python
movies.to_csv('movies.csv')
```

### Write a python program to display IMDB’s Top rated 100 Indian movies’ data 
(i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file

```python
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
response = requests.get(url)
soup3 = BeautifulSoup(response.text, 'lxml')
```

```python
movies = soup3.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup3.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup3.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup3.select('td.posterColumn span[name=nv]')]
```

#### Links

```python
#The site doesn't have the movie names directly, but has hyperlinks to those particular movies respectively.

links = [a.attrs.get('href') for a in soup3.select('td.titleColumn a')]
links
```

#### Ratings

```python
#The ratings are seen under postercolumns of <td tag in <span name = ir

ratings = [b.attrs.get('data-value') for b in soup3.select('td.posterColumn span[name=ir]')]
ratings
```

#### Votes

```python
#The votes are seen under postercolumns of <td tag in <span name = nv

votes = [b.attrs.get('data-value') for b in soup3.select('td.posterColumn span[name=nv]')]
votes[0]
```

#### Main code

```python
import re

imdb = []

# Store each item into dictionary (data), then put those into a list (imdb)
for index in range(0, 100):
    # Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', '')) #To seperate 
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    imdb.append(data)

for item in imdb:
    print(item['place'], '-', item['movie_title'], '('+item['year']+') -', 'rating:', item['rating'])
```
