```python
import pandas as pd
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

#### Fetching the URL

```python
#opening the homepage of amazon

driver.get('https://www.amazon.in/')
```

#### Letting the user to search for anything

```python
#Asking the user to input the keywords.

user_inp = input('enter here for the products: ')


search_bar = driver.find_element_by_id('twotabsearchtextbox')  #Locating the search bar by id
search_bar.clear()  # CLearing the search bar

search_bar.send_keys(user_inp)  #Entering in the search bar.

time.sleep(2)  #To pause for 2 seconds so as to not cause any strain to their server.

search_button = driver.find_element_by_id('nav-search-submit-button')
search_button.click()
```

## Q. 2

#### Collecting URLs

```python
urls = []


for page in range(0,3):               #for loop for scrapping 3 page

    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')#scraping urls

    for i in url:

        urls.append(i.get_attribute("href"))#appending the urls in product_urls list

    nxt_button=driver.find_element_by_xpath("//li[@class='a-last']//a")#scraping the list of buttons from the page

    nxt_button.click()

    time.sleep(2)
```

```python
len(urls)
```

```python
#The empty lists for appending it later.

brands = []
product_Name = []
rating = []
ratings_count = []
price = []
exchange = []
delivery = []
availability = []
other_details = []
product_url = []

for i in urls:
    driver.get(i)
    time.sleep(1)
    
    #Brand
    
    try:
        if (driver.find_element_by_xpath('//th[@class = "a-color-secondary a-size-base prodDetSectionEntry"]').text) == 'Brand':
            comps = driver.find_element_by_xpath("//td[@class = 'a-size-base prodDetAttrValue']").text
            brands.append(comps)
    except:
        brands.append('-')
        
    #Product_name
    
    try:
        if (driver.find_element_by_xpath('//th[@class = "a-color-secondary a-size-base prodDetSectionEntry"]').text) == 'Model Name':
            names = driver.find_element_by_xpath("//td[@class = 'a-size-base prodDetAttrValue']").text
            product_Name.append(names)
    except:
        product_Name.append('-')
            
    #ratings
    
    try:
        star1 = driver.find_element_by_xpath("//span[@class = 'a-size-medium a-color-base']").get_attribute('innerHTML').split(' ')[0]
        rating.append(star1)
    except:
        rating.append('-')
        
    #ratings_count
    
    try:
        count = driver.find_element_by_id("acrCustomerReviewText")
        ratings_count.append(count.text)
    except:
        ratings_count.append('-')
        
    #Price
    
    try:
        cost = driver.find_element_by_id("priceblock_ourprice")
        price.append(cost.text)
    except:
        price.append('-')
        
    #exchange
    
    try:
        ex = driver.find_element_by_xpath('//*[@id="maxBuyBackDiscountSection"]/span/span').text
        exchange.append(ex)
    except:
        exchange.append('-')
        
    #delivery
    
    try:
        deli = driver.find_element_by_id('ddmDeliveryMessage').text.split('.')[0]
        delivery.append(deli)
    except:
        delivery.append('-')
        
    #Availability
    
    try:
        avail = driver.find_element_by_id('availability').text
        availability.append(avail)
    except:
        availability.append('-')
        
    #Other_details
    
    try:
        det = driver.find_element_by_xpath("//ul[@class = 'a-unordered-list a-vertical a-spacing-mini']").text
        other_details.append(det)
    except:
        other_details.append('-')
        
#driver.find_element_by_id('nav-search-submit-button').click()

#time.sleep(2)

#driver.find_element_by_xpath('//li[@class = "a-last"]/a').click

#time.sleep(2)
```

#### DataFraming the Fetched data into a table.

```python
products = pd.DataFrame({})
products['Brand'] = brands
products['rating'] = rating
products['ratings_count'] = ratings_count
products['Price'] = price
products['exchange offer price'] = exchange
products['delivery estimate time'] = delivery
products['availability status'] = availability
products['Other_details'] = other_details
products['product_url'] = urls
```

```python
prod_urls = []



    

for page in range(0,1):               #for loop for scrapping 1st page

    url=rider.find_elements_by_xpath('//a[@class="_1fQZEK"]')#scraping urls

    for i in url:

        prod_urls.append(i.get_attribute("href"))#appending the urls in product_urls list
```

```python
brand = []
name = []
colour = []
ram = []
storage = []
prim_cam = []
sec_cam = []
disp_size = []
proc = []
bat_cap = []
price = []

for i in prod_urls:
    rider.get(i)
    time.sleep(1)
    
    #Brand
    
    try:
        comp = rider.find_element_by_xpath("//span[@class = 'B_NuCI']").text.split(' ')[0]
        brand.append(comp)
    except:
        brand.append('-')
        
    #Model_Name
    
    try:
        peru = rider.find_element_by_xpath('//tr[3][@class = "_1s_Smc row"]').text.split('\n')[1]
        name.append(peru)
    except:
        name.append('-')
        
    #Colour
    
    try:
        color = rider.find_element_by_xpath('//tr[4][@class = "_1s_Smc row"]').text.split('\n')[1]
        colour.append(color)
    except:
        colour.append('-')
        
    #Ram
    
    try:
        mem = rider.find_element_by_xpath('//li[1][@class = "_21Ahn-"]').text.split('|')[0]
        ram.append(mem)
    except:
        ram.append('-')
        
    #storage
    
    try:
        stor = rider.find_element_by_xpath('//li[1][@class = "_21Ahn-"]').text.split('|')[1]
        storage.append(stor)
    except:
        storage.append('-')
        
    #primary_camera
    
    try:
        p_cam = rider.find_element_by_xpath('//li[3][@class = "_21Ahn-"]').text.split('|')[0]
        prim_cam.append(p_cam)
    except:
        prim_cam.append('-')
        
    #secondary_cam
    
    try:
        s_cam = rider.find_element_by_xpath('//li[3][@class = "_21Ahn-"]').text.split('|')[1]
        sec_cam.append(s_cam)
    except:
        sec_cam.append('-')
        
    #display_size
    
    try:
        disp = rider.find_element_by_xpath('//li[2][@class = "_21Ahn-"]').text
        disp_size.append(disp)
    except:
        disp_size.append('-')
        
    #Processor
    
    try:
        process = rider.find_element_by_xpath('//li[5][@class = "_21Ahn-"]').text
        proc.append(process)
    except:
        proc.append('-')
        
    #Battery capacity
    
    try:
        bat = rider.find_element_by_xpath('//li[4][@class = "_21Ahn-"]').text
        bat_cap.append(bat)
    except:
        bat_cap.append('-')
        
   #Price

    try:
        cost = rider.find_element_by_xpath('//div[@class = "_30jeq3 _16Jk6d"]').text
        price.append(cost)
    except:
        price.append('-')
```

```python
time.sleep(3)

try:
    url_string = rower.current_url
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```

## Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
imager = webdriver.Chrome('chromedriver.exe')
```

#### URL

```python
imager.get('https://www.google.co.in/imghp?hl=en&authuser=0&ogbl')
```

#### Search is left to user to search for anything.

#### ex: Fruits, Cars, Machine Learning.

```python
#Asking the user to input the keywords.

user_inp = input('enter here for the item to search on google: ') 

search_bar = imager.find_element_by_name('q')  #Locating the search bar by id
search_bar.clear()  # CLearing the search bar

search_bar.send_keys(user_inp)  #Entering in the search bar.

time.sleep(2)  #To pause for 2 seconds so as to not cause any strain to their server.

search_button = imager.find_element_by_class_name('Tg7LZd')
search_button.click()
```

#### Scrolling the page for more results.

```python
print("start scrolling to generate more images on the page...")
# 500 time we scroll down by 10000 in order to generate more images on the website
for _ in range(1000):
    imager.execute_script("window.scrollBy(0,10000)")
```

```python
time.sleep(3)
```

#### Collecting urls

```python
images = imager.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
```

```python
img_urls = []
img_data = []
for image in images:
    source= image.get_attribute('src')
    if source is not None:
        if(source[0:4] == 'http'):
            img_urls.append(source)
len(img_urls)
```

```python
img_urls
```

#### 'tqdm' is brought in action to show us the progress bar.

#### 'Urllib' module is the URL handling module for python. It is used to fetch URLs.

```python
import urllib.request
import numpy as np
from tqdm import tqdm
sleeps = [1,0.5,1.5,0.7]
```

#### Downloading 100 images

```python
# urllib saves images into the current notebook saved folder and renames using searched item's name.
for i,link in enumerate(tqdm(img_urls)):
    
    if i >= 100:
        break

    name = user_inp+f'{i}.jpeg'

    urllib.request.urlretrieve(link, name)
    time.sleep(np.random.choice(sleeps))
        
imager.quit()
```
